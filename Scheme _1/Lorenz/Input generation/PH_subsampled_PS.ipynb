```python
import matplotlib
import numpy as np
from numpy import zeros, linspace, pi, cos, array
from ripser import ripser
from persim import plot_diagrams
from matplotlib import pyplot as plt
import pandas as pd
import sys

import random
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm


data = np.loadtxt('RAW_99.6_data(600).dat')

a , b, c = data

club = list(zip(a,b,c))

points = [600,500,400,300,200,100]  ### change according to your requirement


for i in range(len(points)):
    
    size_ran = points[i]
    print(size_ran)
    
    z = random.sample(club, size_ran)
    a1, b1, c1 = zip(*z)
    
    n = len(a1)

    data_r = a1,b1,c1
    
    np.savetxt(f'RAW_99.6_data({n}).dat', data_r)

    print("length of data",len(a1))

    fig=plt.figure()
    ax=fig.gca(projection='3d')
    ax.plot(a1,b1,c1,'.k')

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    ax.legend()

    fig.savefig(f'Lorentz_sampling_Rho=99.6_{n}.png')
    
   
    data = np.loadtxt(f'RAW_99.6_data({n}).dat')
    #print(data)
    fig = plt.figure()
    data1=data.transpose()
    #print(data1)
    plt.title("PERSISTENCE DIAGRAM")
    diagrams = ripser(data1, maxdim=1)['dgms']
    plot_diagrams(diagrams, title="PERSISTENCE DIAGRAM", show=True)
    fig.savefig(f'persistence diagram_Rho=99.6,{n}.png')

    #print(diagrams[1])
    np.savetxt(f'ripser_plot_Rho=99.6_{n}.dat', diagrams[1], fmt='%s')

    fig = plt.figure()
    plot_diagrams(diagrams, title="PERSISTENCE lifetime", lifetime=True)
    fig.savefig(f'persistence lifetime_Rho=99.6,{n}.png')

    #------------------------------------------------------------------------------------------------------------------------------
    ### Dat to Csv 

    data_new = np.loadtxt(f'ripser_plot_Rho=99.6_{n}.dat')
    data_n = diagrams[1]
    #print(data_new)
    one = []
    two = []
    length = len(data_new)
    for i in range(length):
        on = data_new[i][0]
        one.append(on)
        tw = data_new[i][1]
        two.append(tw)

    #print(one)
    #print(two)

    dict = {'birth': one, 'death': two} 
        
    df = pd.DataFrame(dict)
    #print(df)

    # saving the dataframe

    df.to_csv(f'ripser_plot_Rho=99.6_{n}.csv')

    #-----------------------------------------------------------------------------------------------------------------------------


    data = pd.read_csv(f'ripser_plot_Rho=99.6_{n}.csv', header=0, encoding ="ISO-8859-1")
    df=pd.DataFrame(data)
    df = data[-data.isin(['inf'])]
    df['diff']=df['death']-df['birth']
    #print(df)
    
    
    
    
    df.sort_values(["diff"], axis=0, ascending=True, inplace=True)
    
    
    df.insert(0, 'Sample_size', n)
    df.drop(df.columns[df.columns.str.contains('unnamed', case=False)], axis=1, inplace=True)

    
    
    print(df)
    df.to_csv(f'ripser_plot_Rho=99.6_{n}(diff).csv')
  #  np.savetxt(f'ripser_barcode(99.6).dat',df)

    #--------------------------------------------------------------------------------------------------------------------------

    %matplotlib inline

    fig = plt.figure()
    table = pd.read_csv(f"ripser_plot_Rho=99.6_{n}(diff).csv")
    table.head()
    plt.title(f"BARCODE_{n},Rho=99.6")
    plt.xlabel('holes')
    plt.ylabel('lifetime')
    plt.bar(x=np.arange(len(table['diff'])),height=table['diff'])
    fig.savefig(f'Barcode_Rho=99.6,{n}.png')
    print()
    print()

```
